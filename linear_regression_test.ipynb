```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split

train = pd.read_csv('./data/train_data.csv')
train = train.iloc[:, 1:]
train
```

```python
train = train.drop(np.where(train.time == 0)[0].tolist())
train
```

```python
X = train.iloc[:, :17]
y = train.iloc[:, 18]

X_train, X_test, y_train, y_test = train_test_split(X, np.log(y), test_size=0.2)
```

```python
lr = LinearRegression()
lr.fit(X_train.X0.to_numpy().reshape(-1, 1), y_train)
pred = lr.predict(X_test.X0.to_numpy().reshape(-1, 1))
error = mean_absolute_error(pred, y_test)
print(error)
```

```python
for i in range(17):
    lr = LinearRegression()
    lr.fit(X_train.iloc[:, i].to_numpy().reshape(-1, 1), y_train)
    pred = lr.predict(X_test.iloc[:, i].to_numpy().reshape(-1, 1))
    error = mean_absolute_error(y_test, pred)
    
    print(f"{str(i)}-{error}")
```

```python
##

lr = LinearRegression()
lr.fit(X_train.apply(sum, axis=1).to_numpy().reshape(-1, 1), y_train)
pred = lr.predict(X_test.apply(sum, axis=1).to_numpy().reshape(-1, 1))
error = mean_absolute_error(pred, y_test)
print(error)
```
