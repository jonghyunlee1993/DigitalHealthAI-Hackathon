```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("./data/train_data.csv")
data = data.iloc[:, 1:]
data
```

```python
remove_index = np.where(data.time == 0)[0].tolist()
remove_index.extend(np.where(data.time >= 100)[0].tolist())

# print(remove_index)

data = data.drop(remove_index)
data
```

```python
data.iloc[:, 17]
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=100, random_state=0)
kmeans.fit(data.iloc[:, [4, 6, 10, 13, 14]])
pred = kmeans.predict(data.iloc[:, [4, 6, 10, 13, 14]])
```

```python
from sklearn.manifold import TSNE

model = TSNE(learning_rate=100)
transformed = model.fit_transform(data.iloc[:, [4, 6, 10, 13, 14]])

xs = transformed[:,0]
ys = transformed[:,1]
plt.scatter(xs, ys, c=pred)

plt.show()
```

```python
plt.scatter(xs, ys, c=data.Action)

plt.show()
```

```python
label_0 = []
label_1 = []

for action, my_pred in zip(data.Action.to_list(), pred.tolist()):
    if action == 0:
        label_0.append(my_pred)
    else:
        label_1.append(my_pred)
```

```python
from collections import Counter
import seaborn as sns

label_0_count = dict(Counter(label_0))
label_1_count = dict(Counter(label_1))

# label_0_count
```

```python
count_result = pd.DataFrame([label_0_count, label_1_count], index=['label_0', 'label_1'])
count_result.T.fillna(0).apply(sum, axis=0)
```

```python
temp = count_result.T.sort_index()
temp
```

```python
pd.options.display.max_rows=999

temp = temp.fillna(0)
temp['fin'] = temp.apply(lambda x: 0 if x['label_0'] > x['label_1'] else 1, axis=1)
temp
```

```python
label_0_index = temp[temp['fin'] == 0].index.to_list()
label_1_index = temp[temp['fin'] == 1].index.to_list()
```

```python
test_data = pd.read_csv("./data/test_data.csv")
test_data = test_data.iloc[:, 1:]
test_data
```

```python
y_pred = kmeans.predict(test_data.iloc[:, [4, 6, 10, 13, 14]])
y_pred
```

```python
pred_label = []

for i in y_pred:
    if i in label_0_index:
        pred_label.append(0)
    else:
        pred_label.append(1)
        
pred_label
```

```python
result = pd.DataFrame(zip(list(range(287)), pred_label), columns=['Title', 'action'])
result.to_csv('./result/kmeans_result.csv', index=False)
result
```
