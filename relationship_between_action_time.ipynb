```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

data = pd.read_csv("./data/whole_data.csv")
data = data.iloc[:, 1:]
data
```

```python
clf = LogisticRegression()
clf.fit(data.Time.to_numpy().reshape(-1, 1), data.Action)

pred = clf.predict(data.Time.to_numpy().reshape(-1, 1)).round(0)

from sklearn.metrics import accuracy_score

accuracy_score(data.Action, pred)
```
